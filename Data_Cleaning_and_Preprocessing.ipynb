# Getting the Dataset

Students' Academic Performance Dataset

```python
import pandas as pd
df = pd.read_csv('drive/MyDrive/dataset/Students Academic Performance Dataset.csv')
dataset = df.copy()
print(dataset.shape)
dataset.head(10)
```

# Check the dimensions of the dataset

```python
print(df.shape)
```

# See if there are duplicates occurs

```python
duplicate_rows = dataset[dataset.duplicated()]
print(duplicate_rows)
print()
```

# Removing Duplicates

```python
print(dataset.shape) #let see first how large is the dataset

# Check for duplicate rows
print("There are " + str(dataset.duplicated().sum()) + " duplicated rows") #print("dataset.duplicated().sum())
```

```python
dataset = dataset.drop_duplicates() #drop the duplicates
dataset.shape
```

# Removing irrelevant data - We remove columns that will not be used.

```python
dataset.drop('Relation', inplace=True, axis=1)
dataset.head()
```

# Standardization of letter case

Converting to Lower Case

```python
dataset['gender'] = dataset['gender'].str.lower()
dataset['NationalITy'] = dataset['NationalITy'].str.lower()
dataset['StageID'] = dataset['StageID'].str.lower()
dataset.head(5)
```

Converting to Upper Case

```python
dataset['PlaceofBirth'] = dataset['PlaceofBirth'].str.upper()
dataset['GradeID'] = dataset['GradeID'].str.upper()
dataset['SectionID'] = dataset['SectionID'].str.upper()
dataset['Topic'] = dataset['Topic'].str.upper()
dataset['Semester'] = dataset['Semester'].str.upper()
dataset['ParentAnsweringSurvey'] = dataset['ParentAnsweringSurvey'].str.upper()
dataset['ParentschoolSatisfaction'] = dataset['ParentschoolSatisfaction'].str.upper()
dataset['Class'] = dataset['Class'].str.upper()
dataset.head(5)
```

Converting to Integer

```python
dataset['raisedhands'] = dataset['raisedhands'].astype(int)
dataset['VisITedResources'] = dataset['VisITedResources'].astype(int)
dataset['AnnouncementsView'] = dataset['AnnouncementsView'].astype(int)
dataset['Discussion'] = dataset['Discussion'].astype(int)
dataset.head(5)
```

Capitalized the First Letter

```python
dataset['StudentAbsenceDays'] = dataset['StudentAbsenceDays'].str.capitalize()
dataset.head(5)
```

# Basic Analysis

```python
# Gender Analysis
gender_counts = df['gender'].value_counts()
print("Gender Distribution:")
print(gender_counts)
print("--------------------------------------------------------")

# Nationality Analysis
nationality_counts = df['NationalITy'].value_counts()
print("\nNationality Distribution:")
print(nationality_counts)
print("--------------------------------------------------------")

# Stage and Grade Level Analysis
stage_counts = df['StageID'].value_counts()
print("\nStage Distribution:")
print(stage_counts)
print("--------------------------------------------------------")

grade_counts = df['GradeID'].value_counts()
print("\nGrade Distribution:")
print(grade_counts)
print("--------------------------------------------------------")

# Topic Analysis
topic_counts = df['Topic'].value_counts()
print("\nTopic Distribution:")
print(topic_counts)
print("--------------------------------------------------------")

# Engagement Analysis
engagement_summary = dataset[['raisedhands', 'VisITedResources', 'AnnouncementsView', 'Discussion']].describe()
print("\nEngagement Summary:")
print(engagement_summary)
print("--------------------------------------------------------")

# Parental Involvement and Satisfaction Analysis
parent_survey_counts = dataset['ParentAnsweringSurvey'].value_counts()
print("\nParent Survey Response:")
print(parent_survey_counts)
print("--------------------------------------------------------")

parent_satisfaction_counts = dataset['ParentschoolSatisfaction'].value_counts()
print("\nParent School Satisfaction:")
print(parent_satisfaction_counts)
print("--------------------------------------------------------")

# Absence Analysis
absence_summary = dataset['StudentAbsenceDays'].value_counts()
print("\nAbsence Summary:")
print(absence_summary)
print("--------------------------------------------------------")

# Class Analysis
class_counts = dataset['Class'].value_counts()
print("\nClass Distribution:")
print(class_counts)
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure and subplots
fig, axes = plt.subplots(5, 2, figsize=(18, 30))

# Gender Distribution
sns.countplot(data=df, x='gender', ax=axes[0, 0])
axes[0, 0].set_title('Gender Distribution')

# Nationality Distribution (Top 10)
nationality_counts[:10].plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Top 10 Nationalities')
axes[0, 1].set_xlabel('Nationality')
axes[0, 1].set_ylabel('Count')

# Stage Distribution
sns.countplot(data=df, x='StageID', ax=axes[1, 0])
axes[1, 0].set_title('Stage Distribution')

# Grade Distribution
grade_counts.plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Grade Distribution')
axes[1, 1].set_xlabel('Grade')
axes[1, 1].set_ylabel('Count')

# Topic Distribution (Top 10)
topic_counts[:10].plot(kind='bar', ax=axes[2, 0])
axes[2, 0].set_title('Top 10 Topics')
axes[2, 0].set_xlabel('Topic')
axes[2, 0].set_ylabel('Count')

# Engagement Summary
sns.boxplot(data=df[['raisedhands', 'VisITedResources', 'AnnouncementsView', 'Discussion']], ax=axes[2, 1])
axes[2, 1].set_title('Engagement Summary')
axes[2, 1].set_ylabel('Count')

# Parent Survey Response
sns.countplot(data=df, x='ParentAnsweringSurvey', ax=axes[3, 0])
axes[3, 0].set_title('Parent Survey Response')

# Parent School Satisfaction
sns.countplot(data=df, x='ParentschoolSatisfaction', ax=axes[3, 1])
axes[3, 1].set_title('Parent School Satisfaction')

# Absence Summary
absence_summary.plot(kind='bar', ax=axes[4, 0])
axes[4, 0].set_title('Absence Summary')
axes[4, 0].set_xlabel('Days Absent')
axes[4, 0].set_ylabel('Count')

# Class Distribution
sns.countplot(data=df, x='Class', ax=axes[4, 1])
axes[4, 1].set_title('Class Distribution')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
